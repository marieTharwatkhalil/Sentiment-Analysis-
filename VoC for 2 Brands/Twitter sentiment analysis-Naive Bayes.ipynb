```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop = stopwords.words("english")
stop2=[".","&",";","`","'",'"',':' ,'@',",",'//t.co/5hTFDwwqM7',]
"]" in stop
```

```python
tweets_=[]
tweets_filtered = []
for tweet in tweet_words:
    #print(tweet)
    for word in tweet:
        if word not in stop and word not in stop2:
            #print(word)
            tweets_.append(word)
            #print(tweets_)
    tweets_filtered.append(tweets_)
        #print(tweets_filtered)
tweets_filtered        
```
